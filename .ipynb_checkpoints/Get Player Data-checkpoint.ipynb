```python
#imports
import urllib.request
import json
from bs4 import BeautifulSoup as bs
import re
```

```python
site = "http://na.op.gg/ranking/ladder/page="
```

```python
# at 580 currently
# Get some names
player_names = set()
for site_num in range(1,9821):
    url = urllib.request.urlopen(site + str(site_num))
    data = url.read()
    parsed_data = bs(data, "lxml")

    player_url = "//na.op.gg/summoner/userName="
    size_player_url = len(player_url)
    for a in parsed_data.find_all('a', href=True):
        potential_url = a['href']
        if(potential_url.startswith("//na.op.gg/summoner/userName=")):
            player_names.add(potential_url[size_player_url:])
    if(site_num%10 == 0):
        print(str(site_num) + ": " + str(len(player_names)))
```

```python
players_file = open("players.txt", "w")
for name in player_names:
    players_file.write(name + "\n")
players_file.close()
```

```python
players_file = open("players.txt", "r")
player_names = players_file.readlines()
player_names = [x.strip() for x in player_names] 
```

```python
player_url = "http://na.op.gg/summoner/userName="
mmr_url = "http://na.op.gg/summoner/ajax/mmr/summonerName="
```

```python
i = 1
players = dict()
for name in player_names:
    print(name, end='')
    try:
        url = urllib.request.urlopen(mmr_url + name)
        data = url.read()
        parsed_data = bs(data, "lxml")
        num = ""
        for item in re.findall('\d+', parsed_data.findAll("td", {"class": "MMR"})[0].text): num += item
        num = int(num)
        players[name] = num
        print(": " + str(num))
    except:
        print(": No MMR")
    i = i+1
```
