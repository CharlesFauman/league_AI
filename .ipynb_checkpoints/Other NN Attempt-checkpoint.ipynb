```python
import numpy as np
import tensorflow as tf

from time import time

from __future__ import division
from __future__ import print_function
from __future__ import unicode_literals
```

```python
train_data = None
dev_data = None
test_data = None
with open('data/train_data.bin', 'rb') as file:
    train_data = np.load(file)
with open('data/dev_data.bin', 'rb') as file:
    dev_data = np.load(file)
with open('data/test_data.bin', 'rb') as file:
    test_data = np.load(file)
if(train_data is None or dev_data is None or test_data is None):
    print("Could not load all data")
```

```python
forward_train_data = np.asarray([np.ndarray.astype(np.append(game[:,0],game[:,1]), dtype = np.float32) for game in train_data])
forward_dev_data = np.asarray([np.ndarray.astype(np.append(game[:,0],game[:,1]), dtype = np.float32) for game in dev_data])
forward_test_data = np.asarray([np.ndarray.astype(np.append(game[:,0],game[:,1]), dtype = np.float32) for game in test_data])

reverse_train_data = np.asarray([np.ndarray.astype(np.append(game[:,1],game[:,0]), dtype = np.float32) for game in train_data])
reverse_dev_data = np.asarray([np.ndarray.astype(np.append(game[:,1],game[:,0]), dtype = np.float32) for game in dev_data])
reverse_test_data = np.asarray([np.ndarray.astype(np.append(game[:,1],game[:,0]), dtype = np.float32) for game in test_data])
```

```python
forward_train_ans = np.column_stack((np.ones(len(forward_train_data)),np.zeros(len(forward_train_data))))
forward_dev_ans = np.column_stack((np.ones(len(forward_dev_data)),np.zeros(len(forward_dev_data))))
forward_test_ans = np.column_stack((np.ones(len(forward_test_data)),np.zeros(len(forward_test_data))))

reverse_train_ans = np.column_stack((np.zeros(len(reverse_train_data)),np.ones(len(reverse_train_data))))
reverse_dev_ans = np.column_stack((np.zeros(len(reverse_dev_data)),np.ones(len(reverse_dev_data))))
reverse_test_ans = np.column_stack((np.zeros(len(reverse_test_data)),np.ones(len(reverse_test_data))))
```

```python
final_train_data = np.concatenate([forward_train_data,reverse_train_data])
final_dev_data = np.concatenate([forward_dev_data,reverse_dev_data])
final_test_data = np.concatenate([forward_test_data,reverse_test_data])
```

```python
final_train_ans = np.concatenate([forward_train_ans,reverse_train_ans])
final_dev_ans = np.concatenate([forward_dev_ans,reverse_dev_ans])
final_test_ans = np.concatenate([forward_test_ans,reverse_test_ans])
```

```python
def weight_variable(shape):
    initial = tf.truncated_normal(shape, stddev=0.01)
    return tf.Variable(initial)


def bias_variable(shape):
    initial = tf.constant(0.0, shape=shape)
    return tf.Variable(initial)

# correct labels
y_ = tf.placeholder(tf.float32, [None, 2])

# input data
x = tf.placeholder(tf.float32, [None, 278])

# network
W_fc1 = weight_variable([278, 100])
b_fc1 = bias_variable([100])
h_fc1 = tf.nn.relu(tf.matmul(x, W_fc1) + b_fc1)

W_fc2 = weight_variable([100, 20])
b_fc2 = bias_variable([20])
h_fc2 = tf.nn.relu(tf.matmul(h_fc1, W_fc2) + b_fc2)

W_fc3 = weight_variable([20, 2])
b_fc3 = bias_variable([2])
y = tf.nn.softmax(tf.matmul(h_fc2, W_fc3) + b_fc3)

# define the loss function
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices=[1]))

# define training step and accuracy
train_step = tf.train.MomentumOptimizer(learning_rate=0.1, momentum=0.9).minimize(cross_entropy)
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

# create a saver
saver = tf.train.Saver()

# initialize the graph
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```python
# train
print("Starting the training...")
start_time = time()
best_accuracy = 0.0

num_epochs = 10
batch_size = 100

indicies = list(range(len(final_train_data)))
for i in range(num_epochs):
    np.random.shuffle(indicies)
    for i in range(0,len(final_train_data)-len(final_train_data)%batch_size,batch_size):
        input_teams = final_train_data[indicies[i:i+batch_size]]
        correct_predictions = final_train_ans[indicies[i:i+batch_size]]
        sess.run(train_step, feed_dict={x: input_teams, y_: correct_predictions})
        train_accuracy = sess.run(accuracy, feed_dict={
        x: input_teams, y_: correct_predictions})
        print("step %d, training accuracy %g" % (i, train_accuracy))
    # validate
    dev_accuracy = sess.run(accuracy, feed_dict={
        x: final_dev_data, y_: final_dev_ans})
    if dev_accuracy > best_accuracy:
        saver.save(sess, './model/fc_best')
        best_accuracy = dev_accuracy
        print("Validation accuracy improved: %g. Saving the network." % dev_accuracy)
    else:
        saver.restore(sess, './model/fc_best')
        print("Validation accuracy was: %g. It was better before: %g. " % (dev_accuracy, best_accuracy) +
              "Using the old params for further optimizations.")
print("The training took %.4f seconds." % (time() - start_time))

# validate
print("Best dev accuracy: %g" % best_accuracy)

# train
train_accuracy = sess.run(accuracy, feed_dict={
        x: final_train_data, y_: final_train_ans})
print("Current train accuracy: %g" % train_accuracy)
```
